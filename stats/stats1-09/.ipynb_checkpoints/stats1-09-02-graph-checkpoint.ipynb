```python
### Read ToothGrowth.csv via loadtxt()
data = np.loadtxt("data/ToothGrowth.csv", delimiter=",", comments="!")
```

# 点線図　Line graph
歯の長さ（即ち、第2列目のデータ）の点線図を作ってください。
Draw a line graph with the data of tooth length (i.e., 1st column  of `tooth_data`)

```python
### Construct a line graph:
plt.plot(data[:, 1:2])
plt.show()
```

# 棒図　Bar graph
歯の長さ（第2列目のデータ）の棒図を作ってください。
Draw a bar graph with the data of tooth length (i.e., 1st column  of `tooth_data`)

```python
### Construct a bar graph:
plt.bar(data[:, 1:2], data[:, 0:1])
plt.show()
```

# ヒストグラム　histogram
歯の長さ（第2列目のデータ）の棒図を作ってください。
Draw a bar graph with the data of tooth length (i.e., 1st column  of `tooth_data`)

```python
### Construct a histogram
plt.hist(data[:, 1:2])
plt.show()
```
