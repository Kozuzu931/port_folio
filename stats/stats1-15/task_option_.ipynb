```python
import numpy as np
import scipy.stats as ss
import matplotlib.pyplot as plt
```

```python
t_value = np.zeros(10000)
gauss = ss.norm(loc = 10, scale = 1)
# シミュレーション
for i in range(0, 10000):
    sample = ss.t.pdf(x=i, df = 10)
    t_value[i] = sample
```

```python
k=np.arange(start = -6, stop = 6.1, step = 0.1)
#plt.hist(t_value, bins=50,density=True,label="t value")
plt.plot(k,gauss.pdf(x=k) ,label="pdf")
plt.legend()
plt.show()
```
