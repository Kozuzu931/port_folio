### 標本平均の分布 Features of sample mean

```python
import numpy as np
import scipy.stats as ss
from matplotlib import pyplot as plt
```

```python
# 標本平均を格納する配列sample_mean(要素数10000)を用意し，ゼロに初期化
sample_mean = np.zeros(10000)

# 母平均を5， 母標準偏差を2とする
# set the population mean 5 and population standard deviation 2
mean = 5
std = 2
# サンプルサイズを10とする
# sample size is 10
sample_size = 10

# 標本を10000セット作成し，それぞれの標本に対して平均を求める
# generate 10000 samples , and calculate the sample mean
for i in range(0,10000):
    #サンプルサイズが10の 平均5標準偏差2の正規分布の乱数を生成
    # generate the R.V. of N[5,4] 
    sample = ss.norm.rvs(size=sample_size, loc=mean, scale=std) 
    #標本平均を求める
    # calculate the sample mean
    sample_mean[i]=np.mean(sample) 
```

```python
# 標本平均のヒストグラムをプロット
# plot the histgram of sample mean
plt.hist(sample_mean,bins=30,density=True)
plt.xlabel("mean of samples")
plt.ylabel("probability")
plt.show()
```

```python
#求められた標本平均からさらに平均を求める
# calculate the mean of sample mean
m = np.mean(sample_mean)
```

```python
#求められた標本平均の標準偏差を求める
#calculate the standard deviation of sample mean
s = np.std(sample_mean)
# 理論値となる標準偏差を求める
# calculate the standard deviation in theory
s2 = std/np.sqrt(sample_size)
print(s)
print(s2)
```

```python
# 平均m，標準偏差sの正規分布の確率密度関数を求める
# calculate pdf of the normal distribution N[m,s*s]
k=np.arange(2,8,0.01)
f = ss.norm.pdf(x=k,loc=m,scale=s)
```

```python
# 標本平均のヒストグラムと正規分布の確率密度関数の折れ線グラフを重ねてプロット
# plot the histogram of sample mean and the pdf of the normal distribution
plt.hist(sample_mean,bins=30,density=True)
plt.plot(k,f)
plt.xlabel("mean of samples")
plt.ylabel("probability")
plt.show()
```
