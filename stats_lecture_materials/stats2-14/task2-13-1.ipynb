```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
from statsmodels.stats.outliers_influence import variance_inflation_factor # VIFを求めるため
import itertools
```

```python
dat = pd.read_csv("data/baseball.csv", delimiter=",")
print(dat.head())
dat.dropna(inplace=True) #欠損値NaNを除外  Remove missing values
print(dat.head())
```

```python
y_name = ["salary"]
x_name = [ 'homerun', 'BA', 'error']
```

# Independent variable number 1

```python
for i, j in itertools.product(y_name, x_name):
    # モデルの設定 Construct a linear regression model
    model = smf.ols(formula = "{y} ~ {x}".format(y=i, x=j), data = dat)
    # 回帰分析の実行 Execute linear  regression 
    results = model.fit()
    # VIFを求める
    col = model.exog.shape[1]# 説明変数の列数
    vifs = [variance_inflation_factor(model.exog, i) for i in range(col)]
    print(i, j, "\n")
    print(results.summary())
    print("AIC", results.aic)
    print("BIC", results.bic)
    print(pd.DataFrame(vifs, index=model.exog_names, columns=["VIF"]), "\n\n")
```

# Independent variable number 2

```python
for i, j in itertools.combinations(x_name, 2):
    # モデルの設定 Construct a linear regression model
    model = smf.ols(formula = "{y} ~ {x1} + {x2}".format(y="salary", x1=i, x2=j), data = dat)
    # 回帰分析の実行 Execute linear  regression 
    results = model.fit()
    # VIFを求める
    col = model.exog.shape[1]# 説明変数の列数
    vifs = [variance_inflation_factor(model.exog, i) for i in range(col)]
    print("{y} = {x1} + {x2} \n".format(y="salary", x1=i, x2=j))
    print(results.summary())
    print("AIC", results.aic)
    print("BIC", results.bic)
    print(pd.DataFrame(vifs, index=model.exog_names, columns=["VIF"]), "\n\n")
```

# Independent variable number 3

```python
# モデルの設定 Construct a linear regression model
model = smf.ols(formula = "salary ~ homerun + BA + error", data = dat)
# 回帰分析の実行 Execute linear  regression 
results = model.fit()
# VIFを求める
col = model.exog.shape[1]# 説明変数の列数
vifs = [variance_inflation_factor(model.exog, i) for i in range(col)]
print("AIC", results.aic)
print("BIC", results.bic)
print(pd.DataFrame(vifs, index=model.exog_names, columns=["VIF"]))
```
