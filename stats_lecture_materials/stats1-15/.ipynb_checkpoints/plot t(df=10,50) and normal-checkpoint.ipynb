```python
import numpy as np
import scipy.stats as ss
import matplotlib.pyplot as plt
```

```python
X = np.arange(-4, 4.1, 0.1)
rand_t_10 = ss.t.pdf(x=X, df=10)
rand_t_50 = ss.t.pdf(x=X, df=50)
review_norm = ss.norm.pdf(x=X, loc=0, scale=1)
```

```python
plt.plot(X,rand_t_10)
plt.plot(X,rand_t_50)
plt.plot(X,review_norm, linestyle="dotted")
plt.ylabel("probabillity")
plt.xlabel("random variable")
plt.legend(["t(df=10)","t(df=50)", "normal"])
```
