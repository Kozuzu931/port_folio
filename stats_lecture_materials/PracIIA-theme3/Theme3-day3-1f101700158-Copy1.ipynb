```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm
```

```python
#東建コーポレーション_2018年
data = pd.read_csv("data/1766_2018.csv", encoding="SHIFT-JIS", skiprows=1)
```

```python
data
```

```python
dates = pd.to_datetime(data["日付"])
start_price = data["始値"]
end_price = data["終値"]
high_price = data["高値"]
low_price = data["安値"]
turnover = data["出来高"]
adjust_price = data["終値調整値"]
```

```python
dates += np.timedelta64(3, "D")
dates -= pd.to_datetime('2018-01-04')
dates /= np.timedelta64(1, "D")
#Translated_date /= np.timedelta64(1, 'D')
#np.array(Translated_date).reshape(-1, 1)
#Translated_date.reshape(-1, 1)
#type(Translated_date[0].dtypes)
dates
```

```python
X = sm.add_constant(dates)
model = sm.OLS(adjust_price, X)
results = model.fit()
a0 = results.params[0]
a1 = results.params[1]
r2 = results.rsquared
```

```python
print(results.summary())
```

```python
min_x = min(dates)
max_x = max(dates)
xx = np.linspace(min_x, max_x, 2)
yy = a0 + a1*xx
plt.plot(xx, yy, label="a0:{0:.1f}, a1:{1:.2f}".format(a0, a1))
plt.plot(dates,adjust_price, label="adjust_price_2018")
plt.legend()
```
