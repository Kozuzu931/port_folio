```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#東建コーポレーション_2018年
data = pd.read_csv("data/1766_2018.csv", encoding="SHIFT-JIS", skiprows=1)
```

```python
data
```

```python
dates = pd.to_datetime(data["日付"])
start_price = data["始値"]
end_price = data["終値"]
high_price = data["高値"]
low_price = data["安値"]
turnover = data["出来高"]
adjust_price = data["終値調整値"]
```

```python
plt.plot(start_price)
plt.plot(end_price)
plt.plot(high_price)
plt.plot(low_price)
plt.plot(turnover)
```

```python
rolling = adjust_price.rolling(10).mean()
```

```python
plt.plot(dates, adjust_price)
plt.plot(dates, rolling)
plt.xlabel("Date")
plt.ylabel("Closing price (adjusted)[yen]")
plt.legend(["1766_tohken", "1766_tohken_rolling_mean"])
```
