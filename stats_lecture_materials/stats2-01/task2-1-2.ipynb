```python
import scipy.stats as ss
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data/stats2_1_1.csv", comment="#")
```

```python
data
```

```python
data_CrimeRate = data["CrimeRate"]
data_ExpendittureYear0 = data["ExpenditureYear0"]
```

```python
ss.pearsonr(data_CrimeRate, data_ExpendittureYear0)
```

```python
ax = plt.subplot(111)
ax.plot(data_CrimeRate)
ax.plot(data_ExpendittureYear0)
```

```python
for i in data:
    print(ss.pearsonr(data["CrimeRate"], i))
```
