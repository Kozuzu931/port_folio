```python
import pandas as pd
import numpy as np
import sqlite3
```

# Connecting to DB Tokyo_weather

```python
conn = sqlite3.connect("Tokyo_weather.db")
cur = conn.cursor()
data = pd.read_sql_query("select * from Tokyo", con=conn)
```

# Preview data from DB

```python
data
```

# Prossesing temperature difference

```python
abs_temp_diff = {i:abs(data.気温[i] - data.気温[i - 24]) for i in range(24, len(data.id)-1)}
abs_temp_diff = sorted(abs_temp_diff.items(), key=lambda x:x[1])
abs_temp_diff = abs_temp_diff[::-1]
abs_temp_diff = abs_temp_diff[:10]
pre_result = pd.DataFrame(abs_temp_diff, columns=[("id"), ("気温差（絶対値）")])
```

```python
abs_temp_diff
```

# Create DataFrame of target terms

```python
target_terms = pd.DataFrame(["{month:.0f}/{date:.0f} {time:.0f}:00".format(month=data.iloc[i[0]]["月"], date=data.iloc[i[0]]["日"], time=data.iloc[i[0]]["時"]) for i in abs_temp_diff], columns=["前日との気温差(絶対値)"])
```

```python
result = pd.concat([pre_result, target_terms], axis=1)
result
```

# +-2 days temperature

```python
result_area_2 = {str(i[0]):data.loc[i[0]-2:i[0]+2, "気温"] for i in abs_temp_diff}
```
