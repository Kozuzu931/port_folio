````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sqlite3
````

# conneting to DB Tokyo_weather

````python
conn = sqlite3.connect("Tokyo_weather.db")
cur = conn.cursor()
data = pd.read_sql_query("select * from Tokyo", conn)
````

# preview data from DB

````python
data
````

# Define target_index and rainfall for sum of rainfall(mm) per 3 hour 

````python
target_index = [i for i in range(2, len(data), 3)]
rainfall = data["降水量"]
````

# Agriggeting data with id

````python
sum_amount_rainfall_3hour = {i:(rainfall.iloc[i] + rainfall.iloc[i - 1] + rainfall.iloc[i - 2]) for i in target_index}
sum_amount_rainfall_3hour = sorted(sum_amount_rainfall_3hour.items(), key=lambda x:x[1])
sum_amount_rainfall_3hour = sum_amount_rainfall_3hour[::-1]
sum_amount_rainfall_3hour = sum_amount_rainfall_3hour[:10]
sum_amount_rainfall_3hour
````

# Create DataFrame of target terms

````python
target_time = ["{month1:.0f}/{date1:.0f} {hour1:.0f}:00 ~ {month2:.0f}/{date2:.0f} {hour2:.0f}:00".format(month1=data.iloc[i[0]]["月"] , date1=data.iloc[i[0]]["日"], hour1=data.iloc[i[0]]["時"], month2=data.iloc[i[0]-2]["月"] , date2=data.iloc[i[0]-2]["日"], hour2=data.iloc[i[0]-2]["時"]) for i in sum_amount_rainfall_3hour]
````

````python
target_time = pd.DataFrame(target_time, columns=["期間"])
target_time
````

# Conntact acounting data and term data

````python
pre_result = pd.DataFrame(sum_amount_rainfall_3hour, columns=(["id", "三時間合計降水量(mm)"]))
result = pd.concat([pre_result, target_time], axis=1)
result
````

# find a solution

````python
since_sun_shine =[data[(data.id > i) & (data.日照時間 > 0)].iloc[0].id - i for i in result.id]
````

````python
since_sun_shine_ave = np.mean(since_sun_shine)
print("降水後、日が差すまでの平均日数:",since_sun_shine_ave,  "日")
````
